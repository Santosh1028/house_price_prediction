#### Dataset Features Overview 

```python
# 1. CRIM per capital crime rate by town
# 2. ZN proportion of residential land zoned for lots over 25,000 sq.ft.
# 3. INDUS proportion of non-retail business acres per town
# 4. CHAS Charles River dummy variable (= 1 if tract bounds river; 0 otherwise)
# 5. NOX nitric oxides concentration (parts per 10 million)
# 6. RM average number of rooms per dwelling
# 7. AGE proportion of owner-occupied units built prior to 1940
# 8. DIS weighted distances to five Boston employment centers
# 9. RAD index of accessibility to radial highways
# 10.TAX full-value property-tax rate per 10,000 USD
# 11. PTRATIO pupil-teacher ratio by town
# 12. Black 1000(Bk — 0.63)² where Bk is the proportion of blacks by town
# 13. LSTAT % lower status of the population
```

```python
#Adding target variable to dataframe
dataset['PRICE'] = boston_dataset.target 
# Median value of owner-occupied homes in $1000s
```

### Spliting and Training the Dataset

```python
X = dataset1.drop(['PRICE'], axis=1)
Y = dataset1['PRICE']
```

```python
X.head(5)
```

```python
Y.head()
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

## Model Training

### Linear Regression 

```python
lr=LinearRegression()
lr.fit(X_train, Y_train)
```

```python
lr.score(X_train, Y_train)
```

### Support Vector Regression

```python
from sklearn.svm import SVR
svr_reg = SVR(kernel = 'linear')
svr_reg.fit(X_train, Y_train)
```

```python
svr_reg.score(X_train, Y_train)
```

### Decision Tree Regression

```python
from sklearn.tree import DecisionTreeRegressor
dt_reg = DecisionTreeRegressor(random_state = 0)
dt_reg.fit(X_train, Y_train)
```

```python
dt_reg.score(X_train, Y_train)
```

### Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
rfr_reg = RandomForestRegressor(n_estimators = 10, random_state = 0)
rfr_reg.fit(X_train, Y_train)
```

```python
rfr_reg.score(X_train, Y_train)
```

### XGBOOST 

```python
from xgboost import XGBRegressor
from sklearn import metrics
```

```python
xgboost=XGBRegressor()
xgboost.fit(X_train, Y_train)
```

```python
xgboost.score(X_train, Y_train)
```

### Model Evaluation

```python
#accuracy for prediction on training data
training_data_prediction = xgboost.predict(X_train)
```

```python
# R squared error
score_1 = metrics.r2_score(Y_train, training_data_prediction)

# Mean Absolute Error
score_2 = metrics.mean_absolute_error(Y_train, training_data_prediction)

print("R squared error : ", score_1)
print('Mean Absolute Error : ', score_2)
```
